```python
import os
import pandas as pd
```

```python
mypath = 'dataset_aksara_bataktoba'
```

```python
file_name = []
tag = []
full_path = []
for path, subdirs, files in os.walk(mypath):
  for name in files:
    if name.endswith('.jpg'):
      full_path.append(os.path.join(path,name))
      tag.append(path.split('/')[-1])
      file_name.append(name)
```

```python
aksbatak_dataframe = pd.DataFrame({'path':full_path,'file_name':file_name,'tag':tag})
aksbatak_dataframe.groupby(['tag']).size()
```

<h1>Splitting Data</h1>

<p>Dilakukan pembagian dataset ke dalam kategori <i>data train</i> dan <i>data test</i></p>

```python
from sklearn.model_selection import train_test_split
```

```python
X = aksbatak_dataframe['path']
y = aksbatak_dataframe['tag']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0,stratify=y)
```

```python
df_path_train = pd.DataFrame({'path':X_train,'tag':y_train})
df_path_test = pd.DataFrame({'path':X_test,'tag':y_test})
```
